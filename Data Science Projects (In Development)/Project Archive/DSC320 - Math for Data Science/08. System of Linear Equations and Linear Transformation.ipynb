## 1. Solving a System of Linear Equations
Use Python to solve the following system of equations. 

```python
"""
Code Description:
    - Solving the system of equations by representing the systrem in matrix form Ax = b

Code Breakdown:
    - 1st Part: Defining the matrix A  
    - 2nd Part: Defining the constant vector b
    - 3rd Part: Solving for the system of equations via "linalg.solve()"
        - Function is used to find the solution vector for the given system of equations
        - Output contains values of x1, x2, x3, and x4 that satisfy the equations in the system
"""

# 1st Part:
A = np.array([[27, -10, 4, -29],
              [-16, 5, -2, 18],
              [-17, 4, -2, 20],
              [-7, 2, -1, 8]])

# 2nd Part:
b = np.array([1, -1, 0, 1])

# 3rd Part:
x = np.linalg.solve(A, b)

# Printing results:
x
```

## 2. Visualizing a Linear Transformation

### (a) Find T(x), where x = [2,-2]

```python
"""
Code Description:
    - Applying the linear transformation T to the vector x using matrix A to find T(x) where x = [2,-2]
    - Computing the matrix-vector product
        - T(x) = A * x

Code Breakdown:
    - 1st Part: Defining the matrix A specified in the instructions 
    - 2nd Part: Defining the vector x
    - 3rd Part: Computing T(x) via 'dot()'
"""

# 1st Part:
A = np.array([[np.sqrt(2)/2, -np.sqrt(2)/2],
              [np.sqrt(2)/2, np.sqrt(2)/2]])

# 2nd Part:
x = np.array([2, -2])

# 3rd Part:
Tx = np.dot(A, x)

# Printing results:
Tx
```

### (b) Mimic the code in Section 7.1.3 of Essential Math for Data Science to visualize the linear transformation T

```python
"""
Code Description:
    - Code referenced from textbook pg.251-254
    - Visualizing the effect of linear transformation on a grid of points by plotting original & transformed grid

Code Breakdown:
    - 1st Part: Creating a grid of points via "meshgrid()" 
    - 2nd Part: Plotting the original grid via "scatter()"
        - Keep in mind that the color of the dots corresponds to the sum of the x and y coordinates of each point
    - 3rd Part: Defining the linear transformation matrix
        - Transformation matrix defines how each point in the grid will be transformed
    - 4th Part: Reshaping the grid points for transformation
        - Flattening grid points where each column represwents a point in the grid with its x and y coordinates stacked vertically
    - 5th Part: Applying the transformation matrix to all points
    - 6th Part: Reshaping the transformed points to the original grid shape
        - Transforming grid to be the same format as the original grid
    - 7th Part: Plotting the transformed grid
"""

# 1st Part:
x = np.arange(-10, 10, 1)
y = np.arange(-10, 10, 1)
xx, yy = np.meshgrid(x, y)

# 2nd Part:
plt.scatter(xx, yy, s=10, c=xx+yy)
plt.title('Original Grid')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()

# 3rd Part:
T = np.array([
    [-1, 0],
    [0, -1]
])

# 4th Part:
xy = np.vstack([xx.flatten(), yy.flatten()])

# 5th Part:
trans = T @ xy

# 6th Part:
xx_transformed = trans[0].reshape(xx.shape)
yy_transformed = trans[1].reshape(yy.shape)

# 7th Part:
plt.scatter(xx_transformed, yy_transformed, s=10, c=xx+yy)
plt.title('Transformed Grid')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
